```python
import os 
from glob import glob
import pandas as pd
import numpy as np
```

```python
csvlist = []
day_list =[]
for filename in glob('stock_data/*'):
    names =(filename.split(os.sep)[-1].split('_'))
    csvlist.append('_'.join(names[0:2]))
    if 'day' in names:
        day_list.append('_'.join(names[0:2]))
```
